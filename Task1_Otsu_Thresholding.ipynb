1. Consider an image with 2 objects and a total of 3-pixel values (1 for each object and 
one for the background). Add Gaussian noise to the image. Implement and test Otsuâ€™s 
algorithm with this image. 

Importing Libraries

```python
# Importing Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

ðŸ§ª Otsuâ€™s Thresholding on Noisy Images for Object Segmentation

```python
# Load original RGB image
image = cv2.imread("Images/input.jpg")  # BGR
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Add Gaussian noise (to RGB)
noise = np.random.normal(0, 10, image_rgb.shape).astype(np.int16)
noisy_image = np.clip(image_rgb.astype(np.int16) + noise, 0, 255).astype(np.uint8)

# Save noisy image
cv2.imwrite("Output/Task1/noisy_image.png", cv2.cvtColor(noisy_image, cv2.COLOR_RGB2BGR))

# Convert noisy image to grayscale for Otsu's method
gray = cv2.cvtColor(noisy_image, cv2.COLOR_RGB2GRAY)

# Apply Otsuâ€™s thresholding
_, otsu_result = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Save output
cv2.imwrite("Output/Task1/otsu_segmented.png", otsu_result)
```

Results Displaying

```python

# Display results
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(image_rgb)
plt.axis("off")
plt.subplot(1, 3, 2)
plt.title("Noisy Image")
plt.imshow(noisy_image)
plt.axis("off")
plt.subplot(1, 3, 3)
plt.title("Otsu Threshold")
plt.imshow(otsu_result, cmap='gray')
plt.axis("off")
plt.tight_layout()
plt.show()
```
